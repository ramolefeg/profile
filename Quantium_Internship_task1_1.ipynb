```python
# prompt: print a list of unique product names

unique_product_names = df_transaction_data['PROD_NAME'].unique()
print(unique_product_names)
```

```python
# prompt: remove special characters from product name

import re
df_transaction_data['PROD_NAME'] = df_transaction_data['PROD_NAME'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
```

```python
unique_product_names = df_transaction_data['PROD_NAME'].unique()
print(unique_product_names)
```

```python
# prompt: count words in the column prod_name, order by word count

# Create a new column with the word count for each product name
df_transaction_data['word_count'] = df_transaction_data['PROD_NAME'].str.split().str.len()

# Group the data by product name and count the word count
word_count_by_product = df_transaction_data.groupby('PROD_NAME')['word_count'].sum().reset_index()

# Sort the data by word count in descending order
word_count_by_product = word_count_by_product.sort_values('word_count', ascending=False)

# Print the top 10 products with the highest word count
word_count_by_product
```

```python
# prompt: get transactions where prod_qty is greater or equal to 200
'''

There are two transactions where 200 packets of chips are bought in one transaction
and both of these transactions were by the same customer.

'''

df_transaction_data[df_transaction_data['PROD_QTY'] >= 200]


#Let's see if the customer has had other transactions

df_transaction_data[df_transaction_data['LYLTY_CARD_NBR'] == 226000]

# Remove the customer outlier  with loyalty_card_number  = 226000
df_transaction_data = df_transaction_data[df_transaction_data['LYLTY_CARD_NBR'] != 226000]
```

```python
# prompt: count the number of transactions by date, order by date in ascending order

transaction_count_by_date = df_transaction_data.groupby('DATE')['PROD_QTY'].count().reset_index()
transaction_count_by_date = transaction_count_by_date.sort_values('DATE', ascending=True)
transaction_count_by_date
```

```python
# prompt: plot number of transactions by date

import matplotlib.pyplot as plt

# Get the number of transactions for each date
transaction_count_by_date = df_transaction_data.groupby('DATE')['PROD_NAME'].count()

# Plot the number of transactions by date
plt.figure(figsize=(12, 6))
plt.plot(transaction_count_by_date.index, transaction_count_by_date.values)
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions by Date')
plt.xticks(rotation=45)
plt.show()
```

```python
# prompt: select rows from a datafame which have words chips and chip in their prod_name and put the results into a new dataframe

df_transactions_chips=df_transaction_data[df_transaction_data['PROD_NAME'].str.contains('chips|chip|chp', case=False, regex=True)]

```

```python
df_transactions_chips.describe()
```

```python
df_transactions_chips.info()
df_transactions_chips.describe()
```

```python
# prompt: extract packet size from prod_name  and create a new column called pack_size on df_transactions_chips

import re

df_transactions_chips['pack_size'] = df_transactions_chips['PROD_NAME'].str.extract(r'(\d+)\s*(?:g|kg|ml|l|pcs|pack)', expand=False)
```

```python
df_transactions_chips.info()
```

```python
# prompt: plot a histogram showing number of transactions by pack size

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.hist(df_transactions_chips['pack_size'], bins=10, edgecolor='black')
plt.xlabel('Pack Size')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions by Pack Size')
plt.show()
```

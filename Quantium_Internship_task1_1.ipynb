# **Data Analysis on Customer Segments**

```python
# total sales by LIFESTAGE and PREMIUM_CUSTOMER and plotting the split by these segments to describe which customer segment contribute most to chip sales.

import matplotlib.pyplot as plt
# Calculate total sales by LIFESTAGE and PREMIUM_CUSTOMER
sales_by_segment = df_merged.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].sum().reset_index()

# Create a pivot table with LIFESTAGE as rows and PREMIUM_CUSTOMER as columns
pivot_table = sales_by_segment.pivot(index='LIFESTAGE', columns='PREMIUM_CUSTOMER', values='TOT_SALES')

# Plot the pivot table as a bar chart
plt.figure(figsize=(10, 6))
pivot_table.plot(kind='bar')
plt.xlabel('LIFESTAGE')
plt.ylabel('Total Sales')
plt.title('Total Sales by LIFESTAGE and PREMIUM_CUSTOMER')
plt.show()
```

```python
# Number of customers by LIFESTAGE and PREMIUM_CUSTOMER

# Calculate the number of customers by LIFESTAGE and PREMIUM_CUSTOMER
customer_count_by_segment = df_merged.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['LYLTY_CARD_NBR'].count().reset_index()

# Create a pivot table with LIFESTAGE as rows and PREMIUM_CUSTOMER as columns
pivot_table = customer_count_by_segment.pivot(index='LIFESTAGE', columns='PREMIUM_CUSTOMER', values='LYLTY_CARD_NBR')

# Print the pivot table
print(pivot_table)
```

```python
# Average number of units per customer by LIFESTAGE and PREMIUM_CUSTOMER

# Calculate the average number of units per customer by LIFESTAGE and PREMIUM_CUSTOMER
avg_units_per_customer = df_merged.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['PROD_QTY'].mean().reset_index()

# Create a pivot table with LIFESTAGE as rows and PREMIUM_CUSTOMER as columns
pivot_table = avg_units_per_customer.pivot(index='LIFESTAGE', columns='PREMIUM_CUSTOMER', values='PROD_QTY')

# Print the pivot table
print(pivot_table)
```

```python
# Average price per unit by LIFESTAGE and PREMIUM_CUSTOMER

# Calculate the average price per unit by LIFESTAGE and PREMIUM_CUSTOMER
avg_price_per_unit = df_merged.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].mean() / df_merged.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['PROD_QTY'].mean()

# Create a pivot table with LIFESTAGE as rows and PREMIUM_CUSTOMER as columns
pivot_table = avg_price_per_unit.unstack()

# Print the pivot table
print(pivot_table)
```

```python
# Perform an independent t-test between mainstream vs premium and budget midage and young singles and couples

import pandas as pd
from scipy import stats

# Load the data
df = pd.read_csv('/content/drive/MyDrive/Quantium Data/merged_data.csv')

# Filter the data
df_filtered = df[(df['LIFESTAGE'].isin (['MIDAGE SINGLES/COUPLES'])) & (df['PREMIUM_CUSTOMER'].isin(['Mainstream', 'Premium', 'Budget']))]
# Perform the independent t-test
t_test_result = stats.ttest_ind(df_filtered[df_filtered['PREMIUM_CUSTOMER'] == 'Mainstream']['TOT_SALES'], df_filtered[df_filtered['PREMIUM_CUSTOMER'].isin (['Premium','Budget'])]['TOT_SALES'])

# Print the results
print("t-statistic:", t_test_result.statistic)
print("p-value:", t_test_result.pvalue)

# Interpret the results
if t_test_result.pvalue < 0.05:
  print("There is a statistically significant difference in the average sales between mainstream and premium customers.")
else:
  print("There is no statistically significant difference in the average sales between mainstream and premium customers.")
```

```python
# look at Mainstream - young singles/couples. For instance, let's find out if they tend to buy a particular brand of chips.

df_merged = df[df['LIFESTAGE'] == 'YOUNG SINGLES/COUPLES']
df_merged = df_merged[df_merged['PREMIUM_CUSTOMER'] == 'Mainstream']
brand_counts = df_merged['brand_name'].value_counts()
print(brand_counts)
```

```python
# Work out of there are brands that Mainstream, young singles/couples prefer more than others.

# Calculate the percentage of purchases for each brand
brand_percentages = brand_counts / brand_counts.sum() * 100

# Print the top 5 brands that Mainstream, young singles/couples prefer
print(brand_percentages.sort_values(ascending=False))
```
